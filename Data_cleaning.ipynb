# Cleaning the Data

01.Correct youth_adult column values

```python
df["youth_adult"] = df["youth_adult"].str.lower().str.strip()

df.loc[df["youth_adult"].str.contains("youth",na=False), "youth_adult"]= "Youth"
df.loc[df["youth_adult"].str.contains("adult", na=False), "youth_adult"]= "Adult"

df.rename(columns={"youth_adult":"age_group"}, inplace=True)
```

02.Handling missing values

```python
#check numer of missing values from each column.
df.isna().sum()
```

```python
#Check missing percentage from each column
df.isna().mean()*100
```

03. Cleaning column names

```python
#We make column names lower case, remove spaces before and after and replace blank spaces with  underscores
df.columns=(
    df.columns
    .str.lower()
    .str.strip()
)
```

```python
df.rename(columns={
    "lu2_combined_rate_of_unemployment_and_time_related_underempl":"lu2_unemployment_underemployment_rate",
    "lu3_combined_rate_of_unemployment_and_potential_labour_force":"lu3_unemployment_potential_lf_rate",
    "lu4_composite_measure_of_labour_underutilization":"lu4_labour_underutilization_rate",
    "neet_rate_youth_not_in_employment_nor_currently_in_education":"youth_neet_rate",
    "labour_force_participation_rate":"lf_participation_rate",
    "median_monthly_earnings_at_main_job":"median_monthly_earnings",
    "out_of_labour_force_inactivity_rate":"out_of_lf_inactivity_rate",
    "time_related_underemployment_rate":"time_underemployment_rate"

},inplace=True, errors="raise")
```
